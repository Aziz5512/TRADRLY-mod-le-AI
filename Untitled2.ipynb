```python
import tensorflow as tf
from tensorflow.keras import layers, optimizers

# 1. Paramètres accélérés
IMG_SIZE = (64, 64)  # Réduit la taille (au lieu de 128)
BATCH_SIZE = 64      # Augmente le batch size
EPOCHS = 8          # Réduit le nombre d'epochs

# 2. Chargement des données (version rapide)
def load_data(subset):
    return tf.keras.preprocessing.image_dataset_from_directory(
        f'./cifake_data/{subset}',
        label_mode='binary',
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        shuffle=(subset == 'train')  # Mélange seulement pour l'entraînement
    )

train_ds = load_data('train')
val_ds = load_data('test')  # On utilise le test set comme val pour aller plus vite

# 3. Modèle CNN léger
model = tf.keras.Sequential([
    layers.Rescaling(1./255),  # Normalisation intégrée

    layers.Conv2D(16, (3, 3), activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

# 4. Optimisation hyperparamètres
model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy'],
    steps_per_execution=10  # Accélère l'entraînement sur GPU
)

# 5. Entraînement rapide
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    verbose=1
)

# Évaluation
test_loss, test_acc = model.evaluate(val_ds)
print(f"\nAccuracy finale : {test_acc:.2%}")
```

```python
# Charger les données de test (si pas déjà fait)
test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    './cifake_data/test',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

# Évaluer
test_loss, test_acc = model.evaluate(test_ds)
print(f"Accuracy sur le test set: {test_acc:.2%}")

# Visualiser quelques prédictions
import matplotlib.pyplot as plt
for images, labels in test_ds.take(1):
    predictions = model.predict(images)
    plt.figure(figsize=(10, 10))
    for i in range(9):
        plt.subplot(3, 3, i+1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(f"Real: {labels[i].numpy()}, Pred: {float(predictions[i]) > 0.5}")
        plt.axis("off")
    plt.show()
```

```python
# Identifier les faux positifs/négatifs
y_true = []
y_pred = []
for images, labels in test_ds:
    y_true.extend(labels.numpy())
    y_pred.extend(model.predict(images).flatten() > 0.5)

from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Prédit')
plt.ylabel('Réel')
plt.show()
```

```python
# Fonction de prédiction sur une nouvelle image
def predict_image(image_path):
    img = tf.keras.preprocessing.image.load_img(
        image_path, target_size=IMG_SIZE
    )
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)  # Ajoute une dimension batch

    prediction = model.predict(img_array)*70
    print(prediction)
    return "réelle" if prediction > 0.5 else "ia"

# Exemple d'utilisation
print(predict_image("/content/ayoub_gaaloul.jpg"))
```

```python
# MobileNetV2 pré-entraîné (ultra-rapide)
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(64, 64, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False  # Ne pas ré-entraîner les poids

model = tf.keras.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(1, activation='sigmoid')
])

# Ré-entraînement express (3 epochs suffisent)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(train_ds, validation_data=val_ds, epochs=3)
```
